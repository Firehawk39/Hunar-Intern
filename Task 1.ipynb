```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
```

```python
df = pd.read_csv(io.BytesIO(uploaded['food_coded.csv']))
```

```python
print("Initial dataset information:")
print(df.info())
print("\nInitial dataset head:")
print(df.head())
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df_mean_filled = df.copy()
df_mean_filled[numeric_cols] = df_mean_filled[numeric_cols].fillna(df_mean_filled[numeric_cols].mean())
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
medians = df[numeric_cols].median()
df_median_filled = df.copy()
df_median_filled[numeric_cols] = df_median_filled[numeric_cols].fillna(medians)
```

```python
df_mode_filled = df.apply(lambda x: x.fillna(x.mode()[0]) if x.mode().size > 0 else x)
```

```python
print("\nSum of null values after filling with mean:")
print(df_mean_filled.isnull().sum())

print("\nSum of null values after filling with median:")
print(df_median_filled.isnull().sum())

print("\nSum of null values after filling with mode:")
print(df_mode_filled.isnull().sum())
```

```python
df_no_duplicates = df.drop_duplicates()
```

```python
print("\nSum of null values after removing duplicates:")
print(df_no_duplicates.isnull().sum())
```

```python
print("\nDataset information after filling missing values with mean:")
print(df_mean_filled.info())
print("\nDataset head after filling missing values with mean:")
print(df_mean_filled.head())

print("\nDataset information after filling missing values with median:")
print(df_median_filled.info())
print("\nDataset head after filling missing values with median:")
print(df_median_filled.head())

print("\nDataset information after filling missing values with mode:")
print(df_mode_filled.info())
print("\nDataset head after filling missing values with mode:")
print(df_mode_filled.head())

print("\nDataset information after removing duplicates:")
print(df_no_duplicates.info())
print("\nDataset head after removing duplicates:")
print(df_no_duplicates.head())
```

```python
df_mean_filled.to_csv('cleaned_mean_filled.csv', index=False)
df_median_filled.to_csv('cleaned_median_filled.csv', index=False)
df_mode_filled.to_csv('cleaned_mode_filled.csv', index=False)
df_no_duplicates.to_csv('cleaned_no_duplicates.csv', index=False)
```
